## Required Transformation Steps for Title Basics:


```python
#Convert the single string of genres from title basics into 2 new tables.

## create a col with a list of genres
basics['genres_split'] = basics['genres'].str.split(',')

exploded_genres = basics.explode('genres_split')

exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
unique_genres
```

```python
exploded_genres = exploded_genres.drop_duplicates()
exploded_genres
```

```python
title_genres = exploded_genres[['titleId', 'genres_split']].copy()
title_genres
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
exploded_genres['genre_id'] = exploded_genres['genres_split'].map(genre_map)
exploded_genres = exploded_genres.drop(columns=['genres_split','genres','genre_id'])
```

```python
exploded_genres.head()
```

```python
genre_lookup = pd.DataFrame({'Genre_Name': genre_map.keys(),
                             'Genre_ID': genre_map.values()})

genre_lookup
```

```python
title_genres = title_genres.drop_duplicates()
title_genres
```

```python
exploded_genres = exploded_genres.drop_duplicates()
exploded_genres
```

```python
ratings.head()
```

```python
ratings = ratings.drop_duplicates()
ratings
```

```python
tmdb = tmdb.drop_duplicates()
tmdb
```

```python
title_genres.rename(columns={'titleId': 'tconst'}, inplace=True)
ratings.rename(columns={'titleId': 'tconst'}, inplace=True)
exploded_genres.rename(columns={'titleId': 'tconst'}, inplace=True)
tmdb.rename(columns={'imdb_id': 'tconst'}, inplace=True)
```

## Saving Tables

```python
import pymysql
pymysql.install_as_MySQLdb()

from sqlalchemy import create_engine
from sqlalchemy_utils import create_database, database_exists
```

```python
connection_str = "mysql+pymysql://root:Charly26!X@localhost/movies"
```

```python
engine = create_engine(connection_str)
create_database(connection_str)
```

```python
## basics
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = exploded_genres['tconst'].fillna('').map(len).max()
title_len = exploded_genres['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'endYear':Float(),
    'runtimeMinutes':Integer()}

exploded_genres.to_sql('title_basics',engine,dtype=df_schema,if_exists='replace',index=False)
```

```python
## ratings
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = ratings['tconst'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {    
    "tconst": String(key_len+1), 
    'averageRating':Float(),
    'numVotes':Integer()}

ratings.to_sql('title_ratings',engine,dtype=df_schema,if_exists='replace',index=False)
```

```python
## tmdb
from sqlalchemy.types import *
## Calculate max string lengths for object columns
key_len = tmdb['tconst'].fillna('').map(len).max()
certification_len = tmdb['certification'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    "revenue": Float(),
    'budget':Float(),
    'certification':Text(certification_len+1)}

tmdb.to_sql('tmdb_data',engine,dtype=df_schema,if_exists='replace',index=False)
```
